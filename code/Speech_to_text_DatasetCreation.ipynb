# We can again reassign the data to a new var and create a split after shiffling!

```python
dataset_new = load_dataset("mozilla-foundation/common_voice_11_0", "en", split="train")
```

```python
### now shuffle and split
# Shuffle the full dataset
dataset_new = dataset_new.shuffle(seed=42)

# Now, select your splits
dataset_10 = dataset_new.select(range(int(0.1 * len(dataset_new))))  # 10% split
```

```python
print(len(dataset_10))
```

```python
dataset_30 = dataset_new.select(range(int(0.3 * len(dataset_new))))  # 30% split
dataset_30
```

```python
import os

# Create a directory inside /content
dataset_dir = '/content/dataset_10'
os.makedirs(dataset_dir, exist_ok=True)

# Save the dataset to the directory
dataset_10.save_to_disk(dataset_dir)

print(f"Dataset saved to: {dataset_dir}")
```

```python
from datasets import load_from_disk

# Load the dataset from the saved directory
dataset_10_loaded = load_from_disk('/content/dataset_10')

# Check how many samples are in the dataset
print(f"Number of samples in dataset_10: {len(dataset_10_loaded)}")
```

```python
#### check size of dataset

# Get the size of the dataset folder
dataset_size = sum(os.path.getsize(os.path.join(root, file)) for root, dirs, files in os.walk('/content/dataset_10') for file in files)

# Convert the size to MB
dataset_size_MB = dataset_size / (1024 * 1024)

print(f"Size of dataset_10: {dataset_size_MB:.2f} MB")
```

```python
#### download to local
import shutil
from google.colab import files

# Path to the saved dataset directory
dataset_dir = '/content/dataset_10'

# Create a zip file of the dataset directory
shutil.make_archive('/content/dataset_10', 'zip', dataset_dir)

# Download the zip file to local machine
files.download('/content/dataset_10.zip')
```
